```python
from pathlib import Path
import pandas as pd
```

```python
import sys
sys.path.append('../baseline/')
```

```python
# originals
data_path_e = Path('../../data/eclipse_clear.json')
data_path_m = Path('../../data/mozilla_mozall.json')

df_json_e = pd.read_json(data_path_e, lines=True)
df_json_m = pd.read_json(data_path_m, lines=True)

display(df_json_e.head(2))
display(df_json_m.head(2))
```

```python
# filtered
data_path_fe = Path('../../data/eclipse_72k.json')
data_path_fm = Path('../../data/mozilla_200k.json')

df_json_fe = pd.read_json(data_path_fe)
df_json_fm = pd.read_json(data_path_fm)

display(df_json_fe.head(2))
display(df_json_fm.head(2))
```

```python
df_json_fe = df_json_fe.drop('description', axis=1)
df_json_fm = df_json_fm.drop('description', axis=1)
display(df_json_fe.head(2))
display(df_json_fm.head(2))
```

```python
print(df_json_m['bug_id'].duplicated().any())
```

```python
merged_df_e = pd.merge(df_json_fe, df_json_e[['bug_id', 'description']], on='bug_id', how='inner')

df_json_m_no_duplicates = df_json_m.drop_duplicates('bug_id')
merged_df_m = pd.merge(df_json_fm, df_json_m_no_duplicates[['bug_id', 'description']], on='bug_id', how='inner')

display(merged_df_e.head())
display(merged_df_m.head())
```

```python
print(merged_df_e.description.isna().unique())
print(merged_df_e.stemmed_description.isna().unique())

print(merged_df_m.description.isna().unique())
print(merged_df_m.stemmed_description.isna().unique())
```

```python
merged_df_m_noempty = merged_df_m[
        merged_df_m['description'].apply(
            lambda x: isinstance(x, str) and len(x.strip()) > 0
        )
    ]
```

```python
merged_df_e.to_json("../../data/new_eclipse.json", orient="records", indent=4)
merged_df_m.to_json("../../data/new_mozilla.json", orient="records", indent=4)
```
