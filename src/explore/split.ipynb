```python

import pandas as pd
import json
# !pip install plotly
import plotly.express as px
```

```python
name = "mozilla_200k"
# name = "eclipse_72k"
df = pd.read_json(f"../../data/llm/{name}.json")
with open(f"../../data/split_{name}.json") as fp:
    ids = json.load(fp)
print(len(ids['tr']))
df.loc[:,'dataset'] = 'none'
for k in ["tr","val","test"]:
    bug_ids = set(ids[k])
    print(len(bug_ids),len(ids[k]))
    df.loc[df['bug_id'].isin(bug_ids),'dataset'] = k
df.sort_values(by="dataset")
df.head()
```
