```python

import pandas as pd
import json
# !pip install plotly
import plotly.express as px
```

```python
name = "mozilla_201k"
# name = "eclipse_72k"
df = pd.read_json(f"../../data/llm/{name}.json")
with open(f"../../data/split_{name}.json") as fp:
    ids = json.load(fp)
df.loc[:,'dataset'] = 'none'
for k in ["tr","val","test"]:
    bug_ids = set(ids[k])
    df.loc[df['bug_id'].isin(bug_ids),'dataset'] = k
df.sort_values(by="dataset")
df.head()
```

```python

fig = px.histogram(df,x="dataset",color="binary_severity",title=name)
# value_counts = df['binary_severity'].value_counts().sort_index()

# # Add annotations for the total count at the top of each bar
# for i, count in enumerate(value_counts):
#     fig.add_annotation(
#         x=i,  # X position of the annotation
#         y=count + 1000,  # Y position of the annotation (slightly above the bar)
#         text=str(count),  # Text of the annotation
#         showarrow=False,  # Don't show an arrow pointing to the bar
#     )
fig.show()
```
