```python
import plotly.express as px
import pandas as pd
import json
import numpy as np
```

```python
df_dico = {}
for dataset in ["eclipse_72k","mozilla_200k"]:
    with open(f"../../data/split_{dataset}.json") as fp:
        split = json.load(fp)
    df = pd.read_json(f"../../data/llm/{dataset}.json")
    df["dataset_type"] = ""
    for s in split:
        df.loc[df["bug_id"].isin(set(split[s])),"dataset_type"] = s
    df1 = df.groupby(["dataset_type","bug_severity"]).agg(len).reset_index()
    df_dico[dataset] = df1
```

```python

for dataset in ["eclipse_72k","mozilla_200k"]:
    total = {e["dataset_type"]:e["bug_id"] for e in df_dico[dataset].groupby("dataset_type").agg(sum).reset_index()[["dataset_type","bug_id"]].to_dict(orient="records")}
    # df_dico[dataset]["perc_tot"] = df_dico[dataset]["bug_id"].apply(lambda x:x['bug_id']/total[x['dataset_type']],axis=1)
    def f(x):
        return np.round(x['bug_id']/total[x['dataset_type']]*100)
    df_dico[dataset]["perc"] = (df_dico[dataset].apply(f, axis=1))
    df_dico[dataset].sort_values("bug_severity",inplace=True)
    display(df_dico[dataset][["dataset_type","bug_severity","perc"]])
```
